```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("CSE_student_performances.csv")
df
```

```python
df.columns
```

```python
df.rename(columns={'Age ':'Age'}, inplace=True)
```

```python
df
```

```python
df['Gender'].str.capitalize().value_counts()
```

```python
df.AcademicPerformance = df['AcademicPerformance'].apply(lambda x: "Good" if x =="Goo" else x)
df.AcademicPerformance.value_counts()
```

```python
df.AcademicPerformance.astype('category').cat.codes
```

```python
df.AcademicPerformance.astype('category')
```

```python
df = df.dropna()
```

```python
df.loc[:, 'AcademicPerformance']= pd.factorize(df.AcademicPerformance)[0]
df.loc[:, 'Gender']= pd.factorize(df['Gender'].str.capitalize())[0]
df.loc[:, 'LikePresentation'] = pd.factorize(df['LikePresentation'].str.capitalize())[0]
df.loc[:, 'FaceChallangesToCompleteAcademicTask'] = pd.factorize(df['FaceChallangesToCompleteAcademicTask'].str.capitalize())[0]
df.loc[:, 'DepressionStatus'] = pd.factorize(df['DepressionStatus'].str.capitalize())[0]
df.loc[:, 'TakingNoteInClass'] = pd.factorize(df['TakingNoteInClass'].str.capitalize())[0]
df.loc[:, 'LikeNewThings'] = pd.factorize(df['LikeNewThings'].str.capitalize())[0]
```

```python
df.loc[:,'NumberOfFriend'] = df['NumberOfFriend'].astype(int)
```

```python
df
```

```python
y=df['AcademicPerformance']
X=df.drop('AcademicPerformance',axis=1)
```

```python
# Для взятого датасета и выбранной целевой переменной решите задачу регрессии. Рассмотрите несколько моделей, определите метрики работы моделей на обучающем и тестовом датасетах. Создайте датасет, который содержит полученные метрики (индексы - модели, столбцы - метрики).
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
X = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8) 
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
model.score(X_test, y_test) 
r2_lr=r2_score(y_test, y_pred)
rmse_lr = mean_squared_error(y_test, y_pred) ** 0.5
y_pred_train = model.predict(X_train)
r2_lr_train=r2_score(y_train, y_pred_train)
rmse_lr_train = mean_squared_error(y_train, y_pred_train) ** 0.5
```

```python
from sklearn.tree import DecisionTreeRegressor

model_tree = DecisionTreeRegressor()
model_tree.fit(X_train, y_train)
y_pred_tree = model_tree.predict(X_test)
model_tree.score(X_test, y_test) 
r2_tree = r2_score(y_test, y_pred_tree)
rmse_tree = mean_squared_error(y_test, y_pred_tree) ** 0.5
y_pred_train = model_tree.predict(X_train)
r2_tree_train=r2_score(y_train, y_pred_train)
rmse_tree_train= mean_squared_error(y_train, y_pred_train) ** 0.5
```

```python
from sklearn.ensemble import RandomForestRegressor

model_forest = RandomForestRegressor()
model_forest.fit(X_train, y_train)
y_pred_forest = model_forest.predict(X_test)
model_forest.score(X_test, y_test) 
r2_forest = r2_score(y_test, y_pred_forest)
rmse_forest = mean_squared_error(y_test, y_pred_forest) ** 0.5
y_pred_train = model_forest.predict(X_train)
r2_forest_train=r2_score(y_train, y_pred_train)
rmse_forest_train = mean_squared_error(y_train, y_pred_train) ** 0.5
```

```python
metrics_df = pd.DataFrame({
    'Модель': ['Linear Regression', 'Decision Tree', 'Random Forest'],
    'R2_train': [r2_lr_train, r2_tree_train, r2_forest_train],
    'R2_test': [r2_lr, r2_tree, r2_forest],
    'RMSE_train': [rmse_lr_train, rmse_tree_train, rmse_forest_train],
    'RMSE_test': [rmse_lr, rmse_tree, rmse_forest]
})
```

```python
metrics_df
```

```python
# Подберите оптимальные параметры моделей п.1. Удалось ли увеличить качество предсказания моделями?
from sklearn.model_selection import GridSearchCV

param_grid_lr = {}

param_grid_tree = {
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

param_grid_forest = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

model_lr = LinearRegression()
model_tree = DecisionTreeRegressor()
model_forest = RandomForestRegressor()

grid_search_lr = GridSearchCV(model_lr, param_grid_lr, cv=5)
grid_search_tree = GridSearchCV(model_tree, param_grid_tree, cv=5)
grid_search_forest = GridSearchCV(model_forest, param_grid_forest, cv=5)

grid_search_lr.fit(X_train, y_train)
grid_search_tree.fit(X_train, y_train)
grid_search_forest.fit(X_train, y_train)

best_params_lr = grid_search_lr.best_params_
best_params_tree = grid_search_tree.best_params_
best_params_forest = grid_search_forest.best_params_

y_pred_lr = grid_search_lr.predict(X_test)
y_pred_tree = grid_search_tree.predict(X_test)
y_pred_forest = grid_search_forest.predict(X_test)

r2_lr_optimized = grid_search_lr.best_estimator_.score(X_test, y_test)
rmse_lr_optimized = mean_squared_error(y_test, y_pred_lr) ** 0.5
r2_tree_optimized = grid_search_tree.best_estimator_.score(X_test, y_test)
rmse_tree_optimized = mean_squared_error(y_test, y_pred_tree) ** 0.5
r2_forest_optimized = grid_search_forest.best_estimator_.score(X_test, y_test)
rmse_forest_optimized = mean_squared_error(y_test, y_pred_forest) ** 0.5

metrics_optimized_df = pd.DataFrame({
    'Модель': ['Linear Regression', 'Decision Tree', 'Random Forest'],
    'R2 (оптимизированная)': [r2_lr_optimized, r2_tree_optimized, r2_forest_optimized],
    'RMSE (оптимизированная)': [rmse_lr_optimized, rmse_tree_optimized, rmse_forest_optimized]
})
```

```python
# Оцените качество работы моделей на основе кросс-валидации.
from sklearn.model_selection import cross_val_score
scaler = StandardScaler()
X = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8) 
model_lin = LinearRegression()
model_lin.fit(X_train, y_train)

scores = cross_val_score(model_lin, X, y, cv=5, scoring='neg_mean_squared_error')
rmse_scores = np.sqrt(-scores)
print("Средняя RMSE на кросс-валидации:", rmse_scores.mean())



```

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor

model_tree = DecisionTreeRegressor()

scores = cross_val_score(model_tree, X, y, cv=5, scoring='neg_mean_squared_error')
rmse_scores = np.sqrt(-scores)
print("Средняя RMSE на кросс-валидации:", rmse_scores.mean())
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor

model_forest = RandomForestRegressor()

scores = cross_val_score(model_forest, X, y, cv=5, scoring='neg_mean_squared_error')
rmse_scores = np.sqrt(-scores)
print("Средняя RMSE на кросс-валидации:", rmse_scores.mean())
```

```python
# Постройте кривые обучения (потерь) - графики зависимости качества работы модели от количества входных данных. Что можно сказать про недообучение и переобучение?
proc=30
list_test=[]
list_train=[]
while proc!=100:
    new_df=df.head(int(len(df)*proc/100))
    y=new_df['AcademicPerformance']
    X=new_df.drop('AcademicPerformance',axis=1)
    scaler = StandardScaler()
    X = scaler.fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8) 
    model_tree = RandomForestRegressor()
    model_tree.fit(X_train, y_train)
    y_pred_tree = model_tree.predict(X_test)
    model_tree.score(X_test, y_test) 
    r2_tree = r2_score(y_test, y_pred_tree)
    rmse_tree = mean_squared_error(y_test, y_pred_tree) ** 0.5
    list_test.append(rmse_tree)
    y_pred_train = model_tree.predict(X_train)
    r2_tree_train=r2_score(y_train, y_pred_train)
    rmse_tree_train= mean_squared_error(y_train, y_pred_train) ** 0.5
    list_train.append(rmse_tree_train)
    proc+=5
    
```

```python
plt.plot(list_test, label='Test')
plt.plot(list_train, label='Train')

plt.title('График для двух списков')
plt.xlabel('Итерации')
plt.ylabel('Оценка')

plt.legend()

plt.show()
```
